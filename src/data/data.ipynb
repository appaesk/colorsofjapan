```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df['Month'] = df['Date'].str.split('/').str[0].astype(int)
df.head()
```

```python
# get Temp_C data for each month for format {1: [1, 2, 3, ...], 2: [4, 5, 6, ..], ...}
def get_monthly_data(df, column='Temp_C'):
    monthly_data = {}
    for month in range(1, 13):
        monthly_data[month] = df[df['Month'] == month][column].tolist()
    return monthly_data
```

```python
monthly_data_tempc = get_monthly_data(df,'Temp_C')
print(monthly_data_tempc)
# Cloud_MinRatio
monthly_data_cloud_min = get_monthly_data(df,'Cloud_MinRatio')
print(monthly_data_cloud_min)
# Precipitation_MM
monthly_data_precip = get_monthly_data(df,'Precipitation_MM')
print(monthly_data_precip)
# Snow_CM
monthly_data_snow = get_monthly_data(df,'Snow_CM')
print(monthly_data_snow)
# Sakura
monthly_data_sakura = get_monthly_data(df,'Sakura')
print(monthly_data_sakura)
# Maple
monthly_data_maple = get_monthly_data(df,'Maple')
print(monthly_data_maple)
# Ginko
monthly_data_ginko = get_monthly_data(df,'Ginko')
print(monthly_data_ginko)
# Hydrangea
monthly_data_hydrangea = get_monthly_data(df,'Hydrangea')
print(monthly_data_hydrangea)
```
